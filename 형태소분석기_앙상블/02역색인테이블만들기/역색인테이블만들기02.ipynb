```python
import os
import pickle
import pandas as pd
import numpy as np
from collections import Counter
import time
import glob
from sklearn.model_selection import train_test_split
import requests
from bs4 import BeautifulSoup as bs
import tensorflow as tf
from tensorflow import keras
from konlpy.tag import Kkma,Komoran, Okt, Hannanum
okt = Okt()
han = Hannanum()
komoran = Komoran()
print(os.getcwd())
with open('01데이터전처리+vocab/new_vocab_content','rb') as f:
    new_vocab_content = pickle.load(f)
with open('01데이터전처리+vocab/new_vocab_title','rb') as f:
    new_vocab_title = pickle.load(f)
with open('01데이터전처리+vocab/형태소앙상블_14000','rb') as f:
    data = pickle.load(f)
```

```python
vocab_content_table = pd.DataFrame({'vocab_content':new_vocab_content})
vocab_title_dic = {}
for i in data['title_nouns']:
    ind = i[0]
    val = i[1]
    if val != '':
        for key,value in val.items():
            if key not in  vocab_title_dic.keys():
                vocab_title_dic[key]=[[ind,value]]
            else:
                vocab_title_dic[key].append([ind,value])
vocab_content_dic ={}
for i in data['content_nouns']:
    ind = i[0]
    val = i[1]
    if val != '':
        for key,value in val.items():
            if key not in  vocab_content_dic.keys():
                vocab_content_dic[key]=[[ind,value]]
            else:
                vocab_content_dic[key].append([ind,value])

locations = pd.DataFrame({'location':[y for x,y in sorted(vocab_title_dic.items(), key = lambda x :-len(x[1]))]})
vocab_title_ = pd.DataFrame({'vocab':[x for x,y in sorted(vocab_title_dic.items(), key = lambda x :-len(x[1]))]})
idf = pd.DataFrame({'idf':[len(y) for x,y in sorted(vocab_title_dic.items(), key = lambda x :-len(x[1]))]})
vocab_title_table = pd.concat([vocab_title_, idf, locations ],axis =1)

locations_con = pd.DataFrame({'location':[y for x,y in sorted(vocab_content_dic.items(), key = lambda x :-len(x[1]))]})
vocab_con_ = pd.DataFrame({'vocab':[x for x,y in sorted(vocab_content_dic.items(), key = lambda x :-len(x[1]))]})
idf_con = pd.DataFrame({'idf':[len(y) for x,y in sorted(vocab_content_dic.items(), key = lambda x :-len(x[1]))]})
vocab_con_table = pd.concat([vocab_con_, idf_con, locations_con ],axis =1)
```

```python
with open('vocab_con_table,'wb') as f:
    pickle.dump(vocab_con_table,f)
with open('vocab_title_table','wb') as f:
    pickle.dump(vocab_title_table,f)
```
