```python
from konlpy.tag import Komoran,Okt, Hannanum
import re
import pickle
import pandas as pd
import numpy as np
from collections import Counter
with open('data14000_코모란','rb') as f:
    data = pickle.load(f)
    data = data.iloc[:,:-2]
with open('vocab_content','rb') as f:
    con_vocab_ori = pickle.load(f)
with open('new_vocab_content','rb') as f:
    con_vocab = pickle.load(f)
komoran = Komoran()
okt = Okt()
hannanum = Hannanum()
titles= data['title']
contents = data['content']
```

#  타이틀

```python
okl = []
kol = []
han = []
for c,i in enumerate(titles):
    try:
        i=re.sub('[^ㄱ-힣 ]',' ',str(i))
        okl.append([x for x in okt.nouns(str(i)) if len(x)>1])
        kol.append([x for x in komoran.nouns(str(i)) if len(x)>1])
        han.append([x for x in hannanum.nouns(str(i)) if len(x)>1])
    except:
        pass
    print(c/len(contents))
okl_flatten = []
kol_flatten = []
han_flatten = []
all_flatten = []
all_l = okl + kol + han 
for i in all_l:
    all_flatten+=i
for x in okl:
    okl_flatten+=x
for x in kol:
    kol_flatten+=x
for x in han:
    han_flatten+=x
sorted_all = sorted(Counter(all_flatten).items(), key = lambda x : -x[1])

okt_gt4 = [x for x,y in Counter(okl_flatten).items() if len(x)>3]
kol_gt4 = [x for x,y in Counter(kol_flatten).items() if len(x)>3]
han_gt4 = [x for x,y in Counter(han_flatten).items() if len(x)>3]

okt_gt4_scattered = []
for i in okt_gt4:
    ko = komoran.nouns(i)
    h = hannanum.nouns(i)
    if len(ko)!=1:
        okt_gt4_scattered+=ko
    if len(h)!=1:
        okt_gt4_scattered+=h
okt_gt4_scattered = [x for x in okt_gt4_scattered if len(x)>1]    

kol_gt4_scattered = []
for i in kol_gt4:
    ot = okt.nouns(i)
    h = hannanum.nouns(i)
    if len(ot)!=1:
        kol_gt4_scattered+=ot
    if len(h)!=1:
        kol_gt4_scattered+=h
kol_gt4_scattered = [x for x in kol_gt4_scattered if len(x)>1]  

han_gt4_scattered = []
for i in han_gt4:
    ot = okt.nouns(i)
    ko = komoran.nouns(i)
    if len(ot)!=1:
        han_gt4_scattered+=ot
    if len(ko)!=1:
        han_gt4_scattered+=ko
han_gt4_scattered = [x for x in han_gt4_scattered if len(x)>1]   
print(len(han_gt4),len(kol_gt4), len(okt_gt4))

vocab = Counter(han_gt4_scattered+kol_gt4_scattered+okt_gt4_scattered+okl_flatten+kol_flatten+han_flatten).keys()
##vocab에서 길이 2-6까지의 단어를 토큰으로 인정
##코모란이 추출한 띄워쓰기가 있는 단어는 okt와 한나눔이 처리했으므로 띄워쓰기 있는 단어는 제거
new_vocab = []
for i in vocab:
    if ' ' not in i:
        if len(i)<7:
            new_vocab.append(i)
print(len(new_vocab))
with open('vocab_title','wb') as f:
    pickle.dump(list(vocab), f)
with open('new_vocab_title','wb') as f:
    pickle.dump(list(new_vocab), f)
```
